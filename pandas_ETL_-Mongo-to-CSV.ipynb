```python
import pandas as pd
```

```python
from pymongo import MongoClient
```

```python
client = MongoClient()
```

```python
client
```

```python
db = client.etl
coll = db.diamonds
```

```python
docs = coll.find()
```

```python
for doc in docs:
    print(doc)
```

```python
df = pd.DataFrame(list(docs)) # convert the docs into dataframe
```

```python
df.info
```

```python
df.head()
```

```python
df.drop(columns=["_id","index"], inplace=True) # drop _id and index columns
```

```python
df.head()
```

```python
df.to_csv("mongo-diamonds.csv", index=False)
```

```python
client.close()
```
